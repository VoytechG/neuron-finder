```python
def batch_generator(batch_size=16, ):
    while True:
        indices = np.random.choice(
            MOVIE_LEN - 16, batch_size, replace=False)
        
        X = np.array([gen.get_x(i, resize_shape=FRAME_NEW_SHAPE)
              for i in indices
             ])
        batch_X = np.concatenate(X, 0).reshape((batch_size,) + X[0].shape)
        
        batch_y = np.array([gen.get_y(i, resize_shape=FRAME_NEW_SHAPE) 
         for i in indices])

        yield (batch_X, batch_y)
```

```python
model = unet3d_simply(input_size=(8,) + FRAME_NEW_SHAPE + (1,), start_filters_no=32)
```

```python
def get_indices_split(indices, data_fraction=1, random_state=(42, 27)):
    if data_fraction != 1:
        np.random.seed(random_state[0])
        np.random.shuffle(indices)
        dataset_size = np.round(len(indices) * data_fraction)
        indices = indices[:data_fraction]
    
    indices_train, indices_test, _, _ = train_test_split(
      indices, indices, test_size=0.25, random_state=random_state[0]
    )
    indices_train, indices_validate, _, _ = train_test_split(
      indices_train, indices_train, test_size=0.25, random_state=random_state[1]
    )
    return indices_train, indices_validate, indices_test

def get_indices_sorted(video_length, temporal_dim):
    div = (4 * 4 * 32) 
    batches = video_length // div 
    indices_all = np.array(range(temporal_dim//2+1, batches*div + temporal_dim//2+1))
    return indices_all

indices_all = get_indices_sorted(MOVIE_LEN, 8)
indices_train, indices_validate, indices_test = get_indices_split(indices_all, data_fraction=0.1)
```
